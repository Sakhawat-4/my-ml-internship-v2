## 2. The Question

**Unit of analysis:** one content page (`content_id`), for a given client.

**The decision:** Out of thousands of pages, which ones should a content
reviewer look at first, given they only have time to review a limited number
each week?

**The action:** A reviewer opens the top-ranked pages and decides whether to
refresh, expand, protect, prune, or simply monitor them.

**The cost of a wrong recommendation:** If the system flags a page that
doesn't actually need attention, the reviewer wastes limited time on it while
a genuinely declining page gets ignored. If it misses a real decliner, that
page keeps losing visibility and traffic unnoticed. Both directions cost
real reviewer time and potential lost search visibility.

**Why data/ML helps here, not just a spreadsheet rule:** A simple hand-written
rule (like "flag any page with impressions above X and no update in 180 days")
treats every signal equally and misses interactions between signals. The
starter pipeline's own results show a learned model finding a meaningfully
better ranking than the hand-written baseline rule — evidence that the
relationships between signals (traffic, position, freshness, engagement) are
more complex than one fixed rule can capture.+

```python
import pandas as pd
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")
print(len(df))
print(df["trend_direction"].value_counts())
```
